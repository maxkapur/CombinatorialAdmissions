```julia
using Plots
```

```julia
m = 25
h = 4

r = 0.6
f = rand(m)
t = 1 .- (r*f + (1-r)*rand(m))

sortorder = sortperm(t)

t[:] = t[sortorder]
f[:] = f[sortorder]

v = zeros(m)

for j in 1:m-1
    v[j] = t[j] * f[j] * prod(1 - f[k] for k in j+1:m)
end

v[m] = t[m] * f[m]

apporder = sortperm(v; rev=true)

pl = plot(size=(600, 600), xlabel="f", ylabel="t")

scatter!(pl, f, t, msw=0, msa=0, ms=12)
annotate!(pl, [(f[apporder[i]], t[apporder[i]], text("$i", :white, 10)) for i in 1:m])
```
